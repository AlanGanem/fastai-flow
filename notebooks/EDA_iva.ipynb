```python
import pandas as pd
import tqdm
import numpy as np
```

```python
data = pd.read_csv(r'../data/external/final_hist_tot.csv', sep = ',', encoding = 'utf-8', index_col = 0)
```

```python
print(data.shape)
print(data.columns)

data.set_index(['Doc.compra','Material']).sort_index()
```

```python
valid_data = data.dropna(subset = ['IVAMIRO'])
print(valid_data.shape)
valid_data.nunique()
```

```python
valid_data.nunique()
```

```python
valid_data.dtypes
```

```python
valid_data.isna().sum()
```

```python
valid_data = valid_data.astype(str)
```

```python
print(valid_data.shape)
```

```python
strip_columns = [
    'Material',
    'Filial',
    'IVAPC',
    'PEP',
    'Fornecedor',
    'Contrato',
    'UF',
    'TpImposto',
    'IVAMIRO'
]
for col in tqdm.tqdm(strip_columns):
    valid_data[col] = valid_data[col].str.strip(' ')
    valid_data[col] = valid_data[col].str.lstrip('0')
```

```python
valid_data = valid_data.drop_duplicates(subset = ['Material','Doc.compra','IVAMIRO'])
valid_data.shape
```

```python
valid_data.nunique()
```

```python
valid_data['OrderType'] = valid_data['Doc.compra'].str[0:2]
#valid_data = valid_data.replace({'':np.nan})
valid_data = valid_data.replace({'nan':''})
```

```python
valid_data.head(100)
```

```python
valid_data['Material'].isna().mean()
```

```python
valid_data = valid_data[valid_data['IVAMIRO'].notna()]
valid_data.shape
```

```python
valid_data.to_csv(r'../data/external/hist_prep.csv')
```
